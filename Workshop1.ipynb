```python
import numpy as np

# Task 1 :
array_range = np.arange(10,50)
print("Array with value ranging from 10 to 49 : ", array_range)

# Task 2 :
matrix_3x3 = np.arange(9).reshape(3,3)
print("3x3 matrix with values from 0 to 8 : ", matrix_3x3)

# Task 3 :
identity_matrix = np.eye(3)
print ("3x3 identity matrix : ", identity_matrix)

# Task 4 :
random_array = np.random.random(30) # create a random array
mean_value = random_array.mean() # calculate the mean
print("Random Array : ", random_array)
print("mean : ", mean_value)

# Task 5 :
array_10x10 = np.random.random((10,10))
min_value = array_10x10.min()
max_value = array_10x10.max()
print("10x10 array : ", array_10x10)
print("min value : ", min_value)
print("max value : ", max_value)

# Task 6 :
zero_array = np.zeros(10)
zero_array[4] = 1 # replace sth element with 1
print("modified array : ", zero_array)

# Task 7 :
arr = [1, 2, 0, 0, 4, 0]
reversed_array = arr[::-1]
print("reversed array : ", reversed_array)

# Task 8 :
array = np.ones((6,6), dtype = int)
array[1:-1, 1:-1] = 0
print("array with 1 on border and 0 on inside : ", array)

# Task 9 :
checkerboard = np.zeros((8,8), dtype = int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("checkerboard pattern : ")
print(checkerboard)
```

```python
import numpy as np

# Given arrays
x = np.array([[1,2], [3,5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# add two arrays
sum_array = x+y
print("the sum is : ", sum_array)

# substract two arrays
diff_array = x-y
print("the difference is : ", diff_array)

# multiply array with any integer
product_array = x*2
print("array x multiplied by 2 is : ",product_array)

# find square of each element in array
square_array = x**2
print("square of each element in x is : ", square_array)

# find dot product between v and w
dot_vw = np.dot(v,w)
print("dot product between v and w is : ", dot_vw)

dot_xy = np.dot(x, y)
print("the dot product between x and y is : ", dot_xy)

# concatenation
# concatenate x and y along rows
concat_xy_rows = np.concatenate((x,y), axis = 0)
print("concatenate x and y along rows : ", concat_xy_rows)

# concatenate x and v
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print("Concatenate x and v:\n", concat_xv)
except ValueError as e:
    print("Error:", e)
# error because the arrays must be of same dimension in order to concatenate
```

```python
# Task 1
import numpy as np
# Define matrices
A = np.array([[3, 4], [7, 8]])

# Calculate the inverse of A
A_inv = np.linalg.inv(A)

# Verify A * A^-1 = I
identity = np.dot(A, A_inv)
print("A * A^-1 = ", identity)

# Task 2
# Define matrix B
B = np.array([[5, 3], [2, 1]])

# Calculate AB and BA
AB = np.dot(A, B)
BA = np.dot(B, A)

# Check if AB != BA
print("AB =\n", AB)
print("BA =\n", BA)
print("Are AB and BA equal?", np.array_equal(AB, BA))

# Task 3
# Calculate (AB)^T and B^T * A^T
AB_T = np.transpose(AB)
BT_AT = np.dot(np.transpose(B), np.transpose(A))

# Verify (AB)^T = B^T * A^T
print("(AB)^T =\n", AB_T)
print("B^T * A^T =\n", BT_AT)
print("Are (AB)^T and B^T * A^T equal?", np.array_equal(AB_T, BT_AT))

# Task 4
# Define the coefficient matrix A and the constant matrix B
A_system = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_system = np.array([-1, -3, 9])

# Solve for X using inverse
A_inv_system = np.linalg.inv(A_system)
X = np.dot(A_inv_system, B_system)
print("Solution of the system [x, y, z]:", X)


```
